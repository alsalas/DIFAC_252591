```python
from Bio import ExPASy
from Bio import SwissProt
# Un códido desde Uniprot
code = 'P11166'
handle = ExPASy.get_sprot_raw(code)
prot = SwissProt.read(handle)
secuencia = prot.sequence
secuencia
```

```python
# Aminoacid composition
import requests as rq
from bs4 import BeautifulSoup
url = "https://web.expasy.org/protscale/pscale/Hphob.Doolittle.html"
page = rq.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
import re
import pandas as pd
#print(soup.prettify())
#https://regexr.com
texto   = soup.find_all('pre')
pattern = "(\w+):.(.\d+.\d+)"
table   = re.findall(pattern, str(texto))
dftable = pd.DataFrame(table)
dftable.columns=['AA','KD']
dftable
```

```python
dictaa = {'Cys': 'C', 'Asp': 'D', 'Ser': 'S', 'Gln': 'Q', 'Lys': 'K',
     'Ile': 'I', 'Pro': 'P', 'Thr': 'T', 'Phe': 'F', 'Asn': 'N', 
     'Gly': 'G', 'His': 'H', 'Leu': 'L', 'Arg': 'R', 'Trp': 'W', 
     'Ala': 'A', 'Val':'V', 'Glu': 'E', 'Tyr': 'Y', 'Met': 'M'}

aa1 = [dictaa[aa] for aa in dftable['AA'].to_list()]
dftable['AminoAcids']= aa1
dftable
```
